```python
# write function to calculate mean value per year for a continent for categories

def calculate_mean_over_time(data, category, continent ):
    #in parenthesis are the arguments
    #create a mask that selects the continent of choice
    mask_continent = data['continent'] == continent #requires that dataframe has a column named continent
    data_continent = data[mask_continent]
    
    # loop through years and calculate mean 
    years = data_continent['year'].unique()
    summary = []
    for year in years:
        mask_year = data_continent['year'] == year
        data_year = data_continent[mask_year]
        value = np.mean(data_year[category])
        summary.append((continent, year, value))
        #summary append you can add anything, it adds on to list in rows
        #() is a tuple and adds new row, [] adds a new list 
        
    summary = pd.DataFrame(summary, columns = ['Continent', 'Year', category])
    #builds the summary but it's local in function
    # need to return the summary 
    return summary 
#nothing happens, the fucntion is just defined
```

# How would you calculate the median over time?

```python
# write function to calculate median value per year for a continent for categories

def calculate_median_over_time(data, category, continent ):
    #in parenthesis are the arguments
    #create a mask that selects the continent of choice
    mask_continent = data['continent'] == continent #requires that dataframe has a column named continent
    data_continent = data[mask_continent]
    
    # loop through years and calculate mean 
    years = data_continent['year'].unique()
    summary = []
    for year in years:
        mask_year = data_continent['year'] == year
        data_year = data_continent[mask_year]
        value = np.median(data_year[category])
        summary.append((continent, year, value))
        #summary append you can add anything, it adds on to list in rows
        #() is a tuple and adds new row, [] adds a new list 
        
    summary = pd.DataFrame(summary, columns = ['Continent', 'Year', category])
    #builds the summary but it's local in function
    # need to return the summary 
    return summary 
#nothing happens, the function is just defined
```

```python
def calculate_statistic_over_time(data, category, continent, statistic): 
    # can also use func instead of statistic 
    #Specify statistic you want, can be np.median or diferent function
    """EXPLAIN YOURSELF witha  doc string using triple quotes
    state what function does
    Args:
        data: a pandas data frame and that includes a column heading 'continent and 'year'
        category: a string that is one of the column headins in data
        continent: also a string that is of the values in the continent column 
        statistic: a fucntion that can operate on a list of numbers eg np.mean
    Return: 
        a pandas data frame with hedsing continent, year, category
        """
    
    #Create a mask that selects the continent of choice
    assert 'continent' in data.columns.values
    assert continent in data['continent'].unique()
    mask_continent = data['continent']==continent #Requires that data frame has a column named 'continent'
    data_continent = data[mask_continent]
    
    #Loop through years and calculate the mean
    years = data_continent['year'].unique() #Requires that data frame has a column named 'year'
    summary = []
    for year in years:
        mask_year = data_continent['year']==year
        data_year = data_continent[mask_year]
        value= statistic(data_year[category])
        summary.append([continent,year,value])
    summary = pd.DataFrame(summary, columns = ["Continent","Year",category])
    return summary

#Ex: calculate_statistic_over_time(df,'pop','asia',np.median)
```
